# Formation Math : Méthode de Monte Carlo
    
Le terme méthode de Monte-Carlo désigne toute méthode visant à calculer une valeur numérique en utilisant
des procédés aléatoires.  
Les méthodes de Monte-Carlo, en référence aux jeux de hasard pratiqués à Monte-Carlo, ont été développées
notamment sous l'impulsion de John von Neumann et Stanislas Ulam, lors de la Seconde Guerre mondiale et des
recherches sur la fabrication de la bombe atomique.  
Ces méthodes sont particulièrement utilisées pour calculer des aires.
    
 </div>



<div class="alert alert-info">
    
## Approximation de $\pi$
    
On considère un disque de rayon R inscrit dans un carré.

La méthode de Monte-Carlo consiste à choisir au hasard un point dans le carré en tirant de façon indépendante ses coordonnées.   
La probabilité que le point appartienne au disque est égale à $\dfrac{\pi R^2}{(2R)^2}=\dfrac{\pi}4$
    
On effectue un grand nombre de fois l'expérience consistant à choisir un point au hasard dans le carré, la fréquence $f = \dfrac{\text{nombre de points situés dans le disque}}{\text{nombre de points total}}$ donne une estimation de la probabilité précédente.

 </div>


```python
from p5 import *
from random import randint
```

```python
# arrêt si on clique avec la souris

nb_points = 0
nb_points_int = 0 
r = 150

def setup() :
    createCanvas(700, 600)
    background(250) 
    fill(255)
    rect(200, 200, 2*r, 2*r)
    circle(350, 350, 2*r)
    
def draw() : 
    global nb_points, nb_points_int     
    nb_points = nb_points + 1
    x = randint(200, 500)
    y = randint(200, 500)
    if dist(x, y, 350, 350) < 150:
        nb_points_int = nb_points_int + 1
        fill("green")
    else:
        fill('red')
    noStroke()
    circle(x, y, 4)
    fill(250)        
    rect(200, 50, 60, 50)
    fill("red")
    text(str(round(4*nb_points_int/nb_points, 4)),  210 , 75)    
    if mouseIsPressed:
        text("le nombre de points est " + str(nb_points),  210 , 125)
        stop()
    

run()
```
