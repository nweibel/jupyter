# Le jeu de la vie

<div class = "alert  alert-info" style="border-left: 3px solid "> 
    
L'objectif de ce projet est de découvrir et présenter **le jeu de la vie**,  un automate cellulaire à deux dimensions, comportant un jeu de règles très simples. Cet automate a pour inventeur John Conway.
    
Le notebook réalisé devra comporter :
- les règles du" jeu de la vie"
- un résumé des connaissances que les différents documents ont permis de recueillir  
- des programmes d’animations respectant les règles du jeu de la vie. On présentera différents types de structures comme des structures oscillantes, des structures stables, des vaisseaux, etc. 
- On pourra également proposer une animation présentant l’évolution d’une situation de départ aléatoire. 
    

Ce projet s'appuie sur la connaissance des tableaux de tableaux.  
Les images et animations seront à réaliser avec le module p5, dont une connaissance préalable est souhaitable. On pourra consulter la page [p5 sur carnets.info.](https://www.carnets.info/jupyter/p5/)
     
</div>
