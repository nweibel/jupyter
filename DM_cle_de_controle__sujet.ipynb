La clé est alors égale à $97−R$.

```python
2550814168025 % 97
59
    
97 - 59
38
```

## Partie 2 : numéro EAN 13
    


<img src = https://capytale2.ac-paris.fr/web/sites/default/files/2021-10-26-14-51-30/ac-paris-1/nathalie.weibel/ean-13.png width = 250>
<figcaption style="text-align: center;">Figure 2 : numéro EAN 13</figcaption>

    
Le dernier chiffre d’un code-barres EAN 13 est une clé de contrôle calculée à partir des douze premiers chiffres selon l’algorithme suivant :
    
— calculer le triple de la somme des chiffres de rang pair, lus de gauche à droite,  
— calculer la somme des chiffres de rang impair, lus de gauche à droite,  
— additionner ces deux sommes partielles,  
— prendre le chiffre des unités de ce total, qu’on appelle $R$,    
— la clé est égale au reste de la division par $10$ de $10−R$.

**Attention** : ici le rang d’un chiffre est sa position dans le nombre en commençant par la position 1 : ce n’est pas l’indice comme dans une liste.
</div>

<div class = "alert alert-block alert-success">
    

**Exemple** :

Voici les douze premiers chiffres d’un code-barres EAN 13 : 9-782940-19961.   
Appliquons l’algorithme pour trouver le treizième chiffre.  
    
— la somme des chiffres de rang pair (ou d’indice impair) est $7 + 2 + 4 + 1 + 9 + 1 = 24$ donc le triple est $24×3 = 72$   
— la somme des chiffres de rang impair (ou d’indice pair) est $9 + 8 + 9 + 0 + 9 + 6 = 41$  
— le total des deux sommes est 72 + 41 = 113  
— le chiffre des unités est $R= 3$  
— puis $10 -R = 10−3 = 7$ et le reste dans la division euclidienne de 7 par 10 est 7.
Donc la clé du code-barres est 7.

Pour en savoir plus sur les codes-barres : https://fr.wikipedia.org/wiki/Code-barres_EAN

## Partie 3 : numéro de carte bleue
    
    


<img src = https://capytale2.ac-paris.fr/web/sites/default/files/2021-10-26-14-51-48/ac-paris-1/nathalie.weibel/visa-electron-800x504.png width = 300>
<figcaption style="text-align: center;">Figure 3 : carte bleue</figcaption>

    
Un numéro de carte bleue est composé de 16 chiffres.   
Il est valide s’il vérifie **la règle de Luhn** dont l’algorithme en langage naturel est le suivant :  
    
— Copier les chiffres du code dans une liste en lisant les chiffres de droite à gauche,  
— Construire à partir de cette liste une nouvelle liste en doublant les chiffres d’indices impairs, et si un double est strictement supérieur à 9 alors le remplacer par la somme de ses chiffres. *Noter que les éléments de la liste ainsi obtenue sont des chiffres*.  
— Si la somme des chiffres de la liste obtenue est un multiple de 10 alors on dit que le numéro est valide.
    
</div>

<div class = "alert alert-block alert-success">
    

**Exemple** :

On considère le numéro de carte suivant : 1258364108472173  
— La liste obtenue en lisant les chiffres du numéro de droite à gauche est `[3, 7, 1, 2, 7, 4, 8, 0, 1, 4, 6, 3, 8, 5, 2, 1]`  
— La liste obtenue par la règle de Luhn est `[3, 5, 1, 4, 7, 8, 8, 0, 1, 8, 6, 6, 8, 1, 2, 2]`   
— La somme des éléments de cette liste est 70 donc ce numéro est valide.
