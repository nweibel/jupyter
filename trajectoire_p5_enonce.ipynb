## Trajectoire de l'image d'un avion.
    
* L'instruction loadImage(chemin) charge une image à partir d'un chemin (url) et crée un objet de classe p5.Image. L'image peut être préchargée en plaçant l'appel à la fonction `loadImage` dans la fonction `preload` à passer en paramètre de la fonction run.
* L'instruction image(p5.Image, x, y) affiche une p5.Image en plaçant le pixel en haut à gauche au point de coordonnées (x, y) dans la zone de dessin.
* L'instruction translate(x, y) permet de faire une translation du repère en plaçant son origine au point de coordonnées (x, y).    
 
</div>

<div class="alert alert-warning">
    
1. Exécuter le code ci-dessous.   
</div>

```python
C = [0, 400]

def preload():
    global img_avion
    img_avion = loadImage("https://capytale2.ac-paris.fr/web/sites/default/files/2022/05-22/14-32-05/avion.png")   


def setup() :
    createCanvas(600, 500)
    background('orange') 

    
def draw() :
    background('orange') 
    translate(C[0], C[1])
    image(img_avion, 0, 0)
    C[0] = C[0] + 1
    C[1] = f(C[0])
    if C[0] > 600:
        C[0], C[1] = [0, 400]
        

run(preload=preload)
```

```python
stop()
```
