## 1. Lien entre la valeur de la grille et l'affichage
    
- Lire le contenu de la fonction `affiche_grille`. 
- Exécuter les 3 cellules ci-dessous et observer l'affichage réalisé.
</div>



````python
from p5 import *
````

````python
def affiche_grille(tab):
    """
    tab est une grille contenant des 0, des 1 ou des 2
    affiche_grille(tab) affiche sous forme de grille les données de tab : 
    les 0 sont représentés par des cases vides, les 1 par une case avec une croix verte
    et les  2 par des cases avec un cercle rouge 
    """
    lignes = len(tab)                    # nombre de lignes
    colonnes = len(tab[0])               # nombre de colonnes
    taille = width / max(lignes, colonnes) # dimension d'une case
  
    for ligne in range(lignes):
        for colonne  in range(colonnes):
            # colonne renseigne sur l'abscisse, et ligne sur l'ordonnée
            fill(255)
            square(colonne * taille, ligne * taille, taille)
            if tab[ligne][colonne] == 1:
                stroke("green") 
                line(colonne * taille + taille//4, ligne * taille + taille//4, \
                     colonne * taille + 3*taille//4, ligne * taille + 3*taille//4)
                line(colonne * taille + 3*taille//4, ligne * taille + taille//4, \
                     colonne * taille + taille//4, ligne * taille + 3*taille//4)        
            elif tab[ligne][colonne] == 2:
                stroke("red")
                circle(colonne * taille + taille//2, ligne * taille + taille//2, taille//2)
            stroke(0)   
````

````python
grille = [[1, 2, 0], [1, 0, 0], [2, 0, 1]]

def setup():                   
    createCanvas(300, 300)     # dimension de la fenetre d'affichage
    stroke(0)                  # couleur des tracés 
    strokeWeight(2)            # épaisseur des tracés

def draw(): 
    affiche_grille(grille)

run()                          
````
