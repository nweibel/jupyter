## Mouvement (relatif) de la tortue
<div class = "alert alert-block alert-info">

|instruction|description|alias|
| :-- | :-- | :-- |
|`forward(pas)` | la tortue avance de la valeur de `pas` |`fd(pas)`|
| `backward(pas)` | la tortue recule de la valeur de `pas`| `bk(pas)`, `back(pas)`  |
| `left(angle)` | la tortue tourne vers la gauche de `angle` degrés | `lt(angle)` |
| `right(angle)` | la tortue tourne vers la droite de `angle` degrés | `rt(angle)`| 
| `setheading(angle)` | la tortue s’oriente dans la direction de `angle` (0 à droite, 90 en haut, etc.) | `seth(angle)`|
| `circle(rayon)` | la tortue trace un cercle dont le rayon est fixé ||
    
</div>

## Mouvement (absolu) de la tortue
<div class = "alert alert-block alert-info">

|instruction|description|alias|
| :-- | :-- | :-- |
|`goto(x, y)` | la tortue se déplace au point de coordonnées `(x ; y)` |`setposition(x, y)`, `setpos(x, y)` |
| `setx(x)` | l’abscisse de la tortue prend la valeur `x`||
| `sety(y)` | l’ordonnée de la tortue prend la valeur `y`||
| ` home()`| la tortue se déplace au point de coordonnées (0 ; 0)||
    
</div>


## Autres méthodes
<div class = "alert alert-block alert-info">

|instruction|description|alias|
| :-- | :-- | :-- |
|`write(texte)`  | écrit la valeur de texte à la position actuelle de la tortue|
 | `write(texte, align='left', font=('Arial', 8, 'normal'))`  | écrit la valeur de `texte` à la position actuelle de la tortue <br> Paramètres optionnels :  `align` peut prendre l'une des valeurs 'left', 'center' ou 'right' <br> `font` est un triplet (fontname, fontsize, fonttype)|
|`distance(x, y)`| renvoie la distance entre la position de la tortue et le point de coordonnées (x, y)||
|`towards(x, y)`|renvoie l'angle de la droite entre la position de la tortue et le point de coordonnées (x, y)||
|`degrees()`|Fixe l'unité de mesure des angles en degrés||
|`radians()`|Fixe l'unité de mesure des angles en radians||
|`begin_fill()` ... `end_fill()` |Encadre les instructions de tracé d'une forme à remplir ||

```python
# réglages initiaux
bgcolor("#00948588")
pencolor("#0081FF")
fillcolor("#0081FF65")
speed(5)

# cinq disques
for i in range(5):
    begin_fill()
    circle(80)
    left(72)
    end_fill()
    

penup()
goto(0, -130)
pendown()
left(36)
pencolor("gold")

# pentagone régulier
for i in range(5):
    fd(152)
    left(72)
    
hideturtle()
done()
```

```python
cote = 30
couleurs = ["yellow", "gold", "orange", "red", "magenta", "purple", "navy", 
            "blue", "cyan", "green", "darkgreen", "sienna", "gray", "black", "white"]

def carre_plein(cote, couleur):
    fillcolor(couleur)
    pendown()
    begin_fill()
    for i in range(4):
        fd(cote)
        left(90)
    end_fill()
    penup()
    
st()
for colonne in range(5):
    for ligne in range(3):
        penup()
        setx(-240 + 120*colonne)
        sety(200 - 50*ligne)
        couleur = couleurs[ligne + 3*colonne]
        carre_plein(cote, couleur)
        fd(cote + 10)
        write(couleur, align='left', font=('courier', 12, 'normal'))
ht()        
done()
```
